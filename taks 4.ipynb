```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Used_Bikes.csv')
df.head(8)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True, axis=0)
```

```python
df.isnull().sum().sum()
```

```python
df['bike_name'].unique()
```

```python
df.groupby('bike_name')['price'].sum()
```

```python
plt.figure(figsize=(100,80))
sns.barplot(x=df['bike_name'].unique(),y=df.groupby('bike_name')['kms_driven'].sum())
plt.xticks(rotation='vertical')
plt.xlabel('Type of the bikes')
plt.title("bikes vs years ")
plt.show()
```

```python
df.boxplot()
plt.xticks(rotation="vertical")
plt.show()
```

```python
plt.boxplot(df['kms_driven'])
plt.show()
```

```python
df.groupby('bike_name')['age'].sum()
```

```python
df['power'].unique()
```

```python
df['power'].value_counts().plot(kind='bar')
plt.xticks(rotation='horizontal')
plt.show()
```

```python
df.groupby('power')['age'].sum().plot(kind='pie')
plt.title('power and kms')
plt.show()
```
