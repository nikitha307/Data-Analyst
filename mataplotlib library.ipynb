```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataFrame = pd.DataFrame({
   "names": ['nikki', 'cherry', 'sreshta', 'sneha', 'sanjay', 'saiteja'],"Marks": [700,600,300,400,500,600]
})
plt.pie(dataFrame["Marks"], labels = dataFrame["names"])
plt.show()
```

```python
import matplotlib.pyplot as plt
   
Name = ['nikki', 'cherry', 'sreshta', 'sneha', 'sanjay', 'saiteja']
Marks= [300,600,300,400,500,600]

plt.bar(Name,Marks)
plt.title('name and marks')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.show()
```
